```python
nrows,ncols=2,3; AX={}
fig = py.figure(figsize=(ncols*6,nrows*4))

x=10**np.linspace(-2,np.log10(0.99))
Q2=10

cnt=1
ax=py.subplot(nrows,ncols,cnt);AX[cnt]=ax
for iset in range(len(ppdf[:100])):
    ax.plot(x,[ppdf[iset].xfxQ2(1,_,Q2) for _ in x],'r-',alpha=0.1)
    ax.set_ylabel(r'$xu^+(x)$',size=30) 

cnt+=1
ax=py.subplot(nrows,ncols,cnt);AX[cnt]=ax
for iset in range(len(ppdf[:100])):
    ax.plot(x,[ppdf[iset].xfxQ2(2,_,Q2) for _ in x],'r-',alpha=0.1)
    ax.set_ylabel(r'$xd^+(x)$',size=30) 

cnt+=1
ax=py.subplot(nrows,ncols,cnt);AX[cnt]=ax
for iset in range(len(ppdf[:100])):
    ax.plot(x,[ppdf[iset].xfxQ2(3,_,Q2) for _ in x],'r-',alpha=0.1)
    ax.set_ylabel(r'$xs^+(x)$',size=30) 
    
cnt+=1
ax=py.subplot(nrows,ncols,cnt);AX[cnt]=ax
for iset in range(len(ppdf[:100])):
    ax.plot(x,[ppdf[iset].xfxQ2(21,_,Q2) for _ in x],'r-',alpha=0.1)
    ax.set_ylabel(r'$xg(x)$',size=30) 
    
for _ in AX:    
    ax=AX[_]
    ax.semilogx()
    ax.set_xlabel(r'$x$',size=30)  
    ax.tick_params(axis='both', which='major', labelsize=20,direction='in')
```

```python
def DeltaG(x,xg):
    x_ = torch.tensor(x)
    g_ = torch.tensor(xg/x)
    DG = torch.trapz(g_,x_)
    return DG
```

```python
DeltaGs = []
for i in range(100):
    deltag = [ppdf[i].xfxQ2(21,_,Q2) for _ in x]
    DeltaGs.append(DeltaG(x,deltag).cpu().detach().numpy())
```

```python
fig = py.figure(figsize=(ncols*5,nrows*4))
ax=py.subplot(nrows,ncols,1);
ax.hist(DeltaGs,bins=50)
```
